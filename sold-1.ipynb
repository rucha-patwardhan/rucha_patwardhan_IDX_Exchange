```python
files = sorted([f for f in os.listdir('raw') if f.startswith('CRMLSSold') and f.endswith('.csv')])
sold = pd.concat([pd.read_csv(f'raw/{f}') for f in files], ignore_index=True)
sold = sold[sold['PropertyType'] == 'Residential']
print(sold.shape)
```

```python
print(sold.shape)
print(sold.dtypes)
```

```python
print(sold['PropertyType'].unique())
```

```python
missing = sold.isnull().sum()
missing_pct = (missing / len(sold) * 100).round(2)
missing_report = pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
missing_report = missing_report[missing_report['missing_count'] > 0].sort_values('missing_pct', ascending=False)
print(missing_report)
```

```python
cols = ['ClosePrice', 'ListPrice', 'OriginalListPrice', 'LivingArea', 'DaysOnMarket', 'BedroomsTotal', 'BathroomsTotalInteger']
print(sold[cols].describe())
```

```python
url = "https://fred.stlouisfed.org/graph/fredgraph.csv?id=MORTGAGE30US"
mortgage = pd.read_csv(url)
print(mortgage.columns.tolist())  # see exact column names
print(mortgage.head())
```

```python
mortgage.columns = ['date', 'rate_30yr_fixed']
mortgage['date'] = pd.to_datetime(mortgage['date'])
mortgage['year_month'] = mortgage['date'].dt.to_period('M')
mortgage_monthly = mortgage.groupby('year_month')['rate_30yr_fixed'].mean().reset_index()
print(mortgage_monthly.tail())
```

```python
sold['year_month'] = pd.to_datetime(sold['CloseDate']).dt.to_period('M')
sold = sold.merge(mortgage_monthly, on='year_month', how='left')
print(sold['rate_30yr_fixed'].isnull().sum())
print(sold[['CloseDate', 'year_month', 'ClosePrice', 'rate_30yr_fixed']].head())
```

```python
sold.to_csv('sold_combined.csv', index=False)
print("Saved sold_combined.csv with mortgage rates!")
```
